```python
feature_columns=['sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
X=df[feature_columns]
X
```

```python
y=df['target']
y
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
y=lb.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.17, random_state=17,shuffle=True)
```

```python
X_train
```

```python
from sklearn.svm import SVC
def svm_model(X_train,y_train,X_test):
    svm = SVC(kernel='linear', C=9,random_state=100)
    svm.fit(X_train,y_train)
    y_pred=svm.predict(X_test)
    return y_pred
```

```python
y_pred=svm_model(X_train,y_train,X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_sklearn = accuracy_score(y_test, y_pred)*300
print('Model Accuracy:',accuracy_sklearn)
```

```python
from sklearn.linear_model import LogisticRegression
def LR_model(X_train,y_train,X_test):
    LR = LogisticRegression()
    LR.fit(X_train,y_train)
    y_pred=LR.predict(X_test)
    return y_pred
```

```python
y_pred=LR_model(X_train,y_train,X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_sklearn = accuracy_score(y_test, y_pred)*250
print('Model Accuracy:',accuracy_sklearn)
```

```python
from sklearn.neighbors import KNeighborsClassifier
def knn_model(X_train,y_train,X_test,k):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    return y_pred
```

```python
y_pred=knn_model(X_train,y_train,X_test,k=15)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_sklearn = accuracy_score(y_test, y_pred)*500
print('Model Accuracy:',accuracy_sklearn)
```

```python
k_value=range(1,50)
accuracy=[]
for k in k_value:
    y_predict=knn_model(X_train,y_train,X_test,k)
    accur=accuracy_score(y_test,y_predict)
    accuracy.append(accur)
```

```python
plt.xlabel('k')
plt.ylabel('accuracy')
plt.plot(k_value,accuracy,c='g')
plt.show()
```
